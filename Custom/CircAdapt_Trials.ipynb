```python
import pandas as pd
import numpy as np
import os

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)

#print(os.listdir("/content"))

file_path = "SAS Database_multiple populations_for Amee Sangani.xlsx"
df = pd.read_excel(file_path)

# clean column names immediately
df.columns = df.columns.str.strip()

print("Shape:", df.shape)
print("\nFirst 20 columns:")
print(df.columns.tolist()[:20])

df.head()
```

```python
print("Has study column?", "study" in df.columns)
print("Has subjectid column?", "subjectid" in df.columns)
print("Has HR column?", "HR" in df.columns)
print("Has Condition_description column?", "Condition_description" in df.columns)
```

```python
# --- Create patient_id ---
df["patient_id"] = df["study"].astype(str) + "_" + df["subjectid"].astype(str)

print("Unique patients:", df["patient_id"].nunique())

# --- Convert key columns to numeric safely ---
cols_to_numeric = [
    "HR", "MAP", "SBP", "DBP",
    "age", "height", "weight", "BMI", "BSA",
    "watts", "rpe"
]

for col in cols_to_numeric:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# --- Fix MAP if missing ---
if "MAP" in df.columns:
    df["MAP"] = df["MAP"].where(
        df["MAP"].notna(),
        df["DBP"] + (df["SBP"] - df["DBP"]) / 3.0
    )

# --- Clean categorical ---
df["Condition_description"] = df["Condition_description"].astype(str).str.strip()
df["study"] = df["study"].astype(str).str.strip()
df["Group"] = df["Group"].astype(str).str.strip()
df["sex"] = df["sex"].astype(str).str.strip()

df.head()
```

```python
required_cols = ["HR", "MAP", "SBP", "DBP"]

df_clean = df.dropna(subset=required_cols).copy()

print("Rows after filtering:", len(df_clean))
print("Patients remaining:", df_clean["patient_id"].nunique())
```

```python
def map_condition(desc):
    desc = str(desc).lower()

    if "supine" in desc and "rest" in desc:
        return "supine_rest"
    elif "upright" in desc and "rest" in desc:
        return "upright_rest"
    elif "rest" in desc:
        return "rest"

    elif "mild" in desc or "stage_1" in desc or "40%vo2" in desc:
        return "mild_exercise"

    elif "moderate" in desc or "60%vo2" in desc:
        return "moderate_exercise"

    elif "peak" in desc or "max" in desc:
        return "peak_exercise"

    elif "hypoxia" in desc or "fio12" in desc or "fio15" in desc or "fio17" in desc:
        return "hypoxia"

    elif "normoxia" in desc or "fio21" in desc:
        return "normoxia"

    else:
        return "other"


df_clean["condition_group"] = df_clean["Condition_description"].apply(map_condition)

df_clean["condition_group"].value_counts()
```

```python
# baseline = rest rows (used to estimate intrinsic parameters)

baseline_df = df_clean[
    df_clean["condition_group"].isin(["rest", "supine_rest", "upright_rest"])
].copy()

print("Baseline rows:", len(baseline_df))
print("Baseline patients:", baseline_df["patient_id"].nunique())
```

```python
baseline_profiles = (
    baseline_df
    .groupby("patient_id")
    .agg({
        "MAP": "median",
        "SBP": "median",
        "DBP": "median",
        "age": "first",
        "BMI": "first",
        "BSA": "first",
        "sex": "first",
        "study": "first",
        "Group": "first",
        "watts": "median",
        "HR": "median"   # only for reference (NOT for fitting)
    })
    .reset_index()
)

print("Baseline profiles:", baseline_profiles.shape)
baseline_profiles.head()
```

```python
simulation_df = df_clean.copy()

print("Simulation rows:", len(simulation_df))
simulation_df.head()
```

```python
def get_condition_modifiers(row):
    cond = row["condition_group"]
    watts = row["watts"] if np.isfinite(row["watts"]) else 0

    # normalize workload (0 to 1)
    load = np.clip(watts / 300.0, 0, 1)

    # base values
    q_mult = 1.0
    p_shift = 0.0

    if cond in ["rest", "supine_rest"]:
        q_mult = 1.0
        p_shift = -2

    elif cond == "upright_rest":
        q_mult = 1.1
        p_shift = 0

    elif cond == "mild_exercise":
        q_mult = 1.5 + 0.5 * load
        p_shift = +5

    elif cond == "moderate_exercise":
        q_mult = 2.0 + 1.0 * load
        p_shift = +10

    elif cond == "peak_exercise":
        q_mult = 3.0 + 2.0 * load
        p_shift = +20

    elif cond == "hypoxia":
        q_mult = 1.3
        p_shift = +5

    else:
        q_mult = 1.2

    return q_mult, p_shift
```

# CircAdapt Loop

```python
import numpy as np
import pandas as pd

MMHG_TO_PA = 133.322

# --- 1) Try importing CircAdapt ---
import circadapt

# --- 2) Utility: normalize categorical info ---
def normalize_sex(value):
    v = str(value).strip().lower()
    if v.startswith("m"):
        return "m"
    if v.startswith("f"):
        return "f"
    return "u"

def parse_group(value):
    try:
        return int(float(str(value).strip()))
    except Exception:
        return 0

# --- 3) Extract model outputs we care about ---
def extract_model_outputs(model):
    p_syart = np.array(model.get("Model.SyArt.p", list), dtype=float) / MMHG_TO_PA
    map_mmhg = float(np.mean(p_syart))
    sbp_mmhg = float(np.max(p_syart))
    dbp_mmhg = float(np.min(p_syart))
    return map_mmhg, sbp_mmhg, dbp_mmhg

# --- 4) Build baseline personalization from patient profile ---
def apply_baseline_personalization(model, baseline_row):
    sex = normalize_sex(baseline_row["sex"])
    age = baseline_row["age"] if np.isfinite(baseline_row["age"]) else 50.0
    bmi = baseline_row["BMI"] if np.isfinite(baseline_row["BMI"]) else 25.0
    bsa = baseline_row["BSA"] if np.isfinite(baseline_row["BSA"]) else 1.9
    group = parse_group(baseline_row["Group"])
    study = str(baseline_row["study"]).lower()

    # size scaling
    bsa_scale = float(np.clip(bsa / 2.0, 0.75, 1.35))
    sex_scale = 1.05 if sex == "m" else (0.95 if sex == "f" else 1.0)

    model["Patch"]["V_wall"][2:] = model["Patch"]["V_wall"][2:] * bsa_scale * sex_scale
    model["Patch"]["Am_ref"][2:] = model["Patch"]["Am_ref"][2:] * bsa_scale

    # passive stiffness from age/BMI
    age_factor = float(np.clip((age - 40.0) / 40.0, -0.5, 1.0))
    bmi_factor = float(np.clip((bmi - 25.0) / 15.0, -0.5, 1.0))
    sf_pas_scale = 1.0 + 0.25 * age_factor + 0.12 * bmi_factor
    k1_base = 8.0 * (1.0 + 0.20 * age_factor)

    model.set("Model.Peri.TriSeg.wLv.pLv0.Sf_pas", float(700.0 * sf_pas_scale), float)
    model.set("Model.Peri.TriSeg.wSv.pSv0.Sf_pas", float(700.0 * sf_pas_scale), float)
    model.set("Model.Peri.TriSeg.wRv.pRv0.Sf_pas", float(650.0 * sf_pas_scale), float)

    model.set("Model.Peri.TriSeg.wLv.pLv0.k1", float(k1_base), float)
    model.set("Model.Peri.TriSeg.wSv.pSv0.k1", float(k1_base), float)
    model.set("Model.Peri.TriSeg.wRv.pRv0.k1", float(k1_base * 0.95), float)

    # baseline contractility by phenotype
    act_scale = {1: 1.05, 2: 0.90, 3: 1.00, 4: 0.95}.get(group, 1.0)

    if "hypoxia" in study:
        act_scale *= 0.98
    if "lvad" in study:
        act_scale *= 0.95

    model.set("Model.Peri.TriSeg.wLv.pLv0.Sf_act", float(60000.0 * act_scale), float)
    model.set("Model.Peri.TriSeg.wSv.pSv0.Sf_act", float(60000.0 * act_scale), float)
    model.set("Model.Peri.TriSeg.wRv.pRv0.Sf_act", float(56000.0 * act_scale), float)

# --- 5) Objective: pressure only (to avoid HR leakage through CO/SV) ---
def pressure_objective(model_map, model_sbp, model_dbp, target_map, target_sbp, target_dbp):
    e_map = (model_map - target_map) / max(target_map, 1.0)
    e_sbp = (model_sbp - target_sbp) / max(target_sbp, 1.0)
    e_dbp = (model_dbp - target_dbp) / max(target_dbp, 1.0)

    return float(
        0.50 * e_map * e_map +
        0.25 * e_sbp * e_sbp +
        0.25 * e_dbp * e_dbp
    )


def estimate_hr_prior(row, baseline_row):
    age = baseline_row["age"]
    watts = row["watts"] if np.isfinite(row["watts"]) else 0

    # baseline HR estimate
    hr_base = 70 - 0.1 * (age - 40)

    # workload contribution
    load = np.clip(watts / 300.0, 0, 1)
    hr_ex = 40 * load

    # condition adjustment
    cond = row["condition_group"]

    cond_adj = 0
    if cond in ["supine_rest"]:
        cond_adj -= 5
    elif cond in ["upright_rest"]:
        cond_adj += 3
    elif cond == "mild_exercise":
        cond_adj += 10
    elif cond == "moderate_exercise":
        cond_adj += 25
    elif cond == "peak_exercise":
        cond_adj += 60
    elif cond == "hypoxia":
        cond_adj += 10

    return np.clip(hr_base + hr_ex + cond_adj, 45, 190)


# --- 6) Simulate one row by searching over t_cycle only ---
def simulate_row_hr(row, baseline_row, beats=8, grid=10):
    q_mult, p_shift = get_condition_modifiers(row)

    # pressure prior from measured baseline MAP
    baseline_map = baseline_row["MAP"] if np.isfinite(baseline_row["MAP"]) else row["MAP"]
    target_map = row["MAP"]
    target_sbp = row["SBP"]
    target_dbp = row["DBP"]

    # simple HR search range
    tc_candidates = np.linspace(0.40, 1.30, grid)

    best_score = np.inf
    best_hr = np.nan
    best_out = None

    for t_cycle in tc_candidates:
        model = circadapt.VanOsta2024()

        # personalize baseline
        apply_baseline_personalization(model, baseline_row)

        # ADD THIS

        # contractility modifier
        sf_mult = 1.0
        
        cond = row["condition_group"]
        
        if cond == "mild_exercise":
            sf_mult = 1.1
        elif cond == "moderate_exercise":
            sf_mult = 1.25
        elif cond == "peak_exercise":
            sf_mult = 1.4
        elif cond == "hypoxia":
            sf_mult = 1.1
        
        # apply to LV
        model.set("Model.Peri.TriSeg.wLv.pLv0.Sf_act",
                  float(60000.0 * sf_mult), float)

        # pressure-flow control
        model.set("Model.PFC.p0", float((baseline_map + p_shift) * MMHG_TO_PA), float)

        # very rough flow target from condition, NOT from measured CO
        q0_base = 5.0 / 60000.0   # 5 L/min baseline
        model.set("Model.PFC.q0", float(q0_base * q_mult), float)

        # candidate cycle length
        model.set("Model.t_cycle", float(t_cycle), float)

        try:
            model.run(beats)
            map_model, sbp_model, dbp_model = extract_model_outputs(model)
            hr_prior = estimate_hr_prior(row, baseline_row)
            hr_model = 60.0 / float(t_cycle)
            
            score = (
                pressure_objective(map_model, sbp_model, dbp_model,
                                   target_map, target_sbp, target_dbp)
                + 0.10 * ((hr_model - hr_prior) / hr_prior) ** 2
            )

            if score < best_score:
                best_score = score
                best_hr = 60.0 / float(t_cycle)
                best_out = {
                    "map_model": map_model,
                    "sbp_model": sbp_model,
                    "dbp_model": dbp_model,
                    "objective": score
                }
        except Exception:
            continue

    return best_hr, best_out
```

```python
def evaluate_group(df, group_col):
    results = []

    for g, sub in df.groupby(group_col):
        if len(sub) < 5:
            continue

        rmse = np.sqrt(np.mean((sub["HR_predicted"] - sub["HR_measured"])**2))
        mae = np.mean(np.abs(sub["HR_predicted"] - sub["HR_measured"]))

        results.append({
            group_col: g,
            "count": len(sub),
            "RMSE": rmse,
            "MAE": mae
        })

    return pd.DataFrame(results).sort_values("RMSE")
```
