First Rudimentary Model

```python
features = [
    "sex",
    "age",
    "height",
    "weight",
    "BMI",
    "BSA",
    "study",
    "Group",
    "Condition_description",
    "watts",
    "rpe"
]

target = "HR"

model_df = df[features + [target]].copy()

model_df.head()
```

```python
model_df.isna().sum()
```

```python
model_df = model_df.dropna(subset=["HR"])

print(model_df.shape)
```

```python
model_df = model_df.drop(columns=["rpe"])
```

```python
model_df["watts"] = pd.to_numeric(model_df["watts"], errors="coerce")
model_df["sex"] = model_df["sex"].replace(".", np.nan)

model_df["sex"] = model_df["sex"].fillna(model_df["sex"].mode()[0])
model_df["watts"] = model_df["watts"].fillna(model_df["watts"].median())
```

```python
model_df.isna().sum()
```

```python
model_df["patient_id"] = df.loc[model_df.index, "patient_id"]
model_df.head()
```

```python
model_df["patient_id"].nunique()
```

```python
from sklearn.model_selection import train_test_split

patients = model_df["patient_id"].unique()

train_patients, test_patients = train_test_split(
    patients,
    test_size=0.2,
    random_state=42
)

train_df = model_df[model_df["patient_id"].isin(train_patients)]
test_df = model_df[model_df["patient_id"].isin(test_patients)]

print("Train rows:", len(train_df))
print("Test rows:", len(test_df))

print("Train patients:", train_df["patient_id"].nunique())
print("Test patients:", test_df["patient_id"].nunique())
```

```python
X_train = train_df.drop(columns=["HR", "patient_id"])
y_train = train_df["HR"]

X_test = test_df.drop(columns=["HR", "patient_id"])
y_test = test_df["HR"]
```

```python
categorical_cols = ["sex", "study", "Group", "Condition_description"]
numeric_cols = [c for c in X_train.columns if c not in categorical_cols]

print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols)
```

```python
X_train = pd.get_dummies(X_train, columns=categorical_cols)
X_test = pd.get_dummies(X_test, columns=categorical_cols)

# align columns in case test set is missing categories
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train_np = X_train.values
X_test_np = X_test.values

y_train_np = y_train.values
y_test_np = y_test.values
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
X_train.dtypes
```

```python
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)

X_train_np = X_train.to_numpy()
X_test_np = X_test.to_numpy()

y_train_np = y_train.to_numpy(dtype=np.float32)
y_test_np = y_test.to_numpy(dtype=np.float32)
```

```python
print(X_train_np.dtype)
print(X_test_np.dtype)
print(y_train_np.dtype)
print(y_test_np.dtype)
```

```python
X_train_t = torch.tensor(X_train_np, dtype=torch.float32)
y_train_t = torch.tensor(y_train_np, dtype=torch.float32).view(-1, 1)

X_test_t = torch.tensor(X_test_np, dtype=torch.float32)
y_test_t = torch.tensor(y_test_np, dtype=torch.float32).view(-1, 1)
```

```python
class HRPredictor(nn.Module):

    def __init__(self, input_dim):
        super().__init__()

        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.model(x)


model = HRPredictor(X_train_t.shape[1])
```

```python
criterion = nn.MSELoss()

optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 300

for epoch in range(epochs):

    model.train()

    preds = model(X_train_t)

    loss = criterion(preds, y_train_t)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch} | Loss {loss.item():.3f}")
```

```python
model.eval()

with torch.no_grad():
    test_preds = model(X_test_t)

    test_loss = criterion(test_preds, y_test_t)

print("Test MSE:", test_loss.item())
```

```python
rmse = torch.sqrt(test_loss)

print("Test RMSE:", rmse.item())
```

Model 2

```python
def map_condition(desc):
    desc = str(desc).lower()

    if "supine" in desc and "rest" in desc:
        return "supine_rest"
    elif "upright" in desc and "rest" in desc:
        return "upright_rest"
    elif "mild" in desc or "stage_1" in desc or "40%vo2" in desc:
        return "mild_exercise"
    elif "moderate" in desc or "60%vo2" in desc:
        return "moderate_exercise"
    elif "peak" in desc:
        return "peak_exercise"
    elif "hypoxia" in desc or "fio12" in desc or "fio15" in desc or "fio17" in desc:
        return "hypoxia"
    elif "normoxia" in desc or "fio21" in desc:
        return "normoxia"
    else:
        return "other"

df["condition_group"] = df["Condition_description"].apply(map_condition)

df["condition_group"].value_counts()
```

```python
reduced_features = [
    "sex",
    "age",
    "BMI",
    "BSA",
    "study",
    "condition_group",
    "watts"
]

target = "HR"

model_df2 = df[reduced_features + [target, "patient_id"]].copy()
model_df2.head()
```

```python
model_df2[target] = pd.to_numeric(model_df2[target], errors="coerce")
model_df2["watts"] = pd.to_numeric(model_df2["watts"], errors="coerce")
model_df2["sex"] = model_df2["sex"].replace(".", np.nan)

model_df2 = model_df2.dropna(subset=[target])

model_df2["sex"] = model_df2["sex"].fillna(model_df2["sex"].mode()[0])
model_df2["watts"] = model_df2["watts"].fillna(model_df2["watts"].median())

model_df2.isna().sum()
```

```python
train_df2 = model_df2[model_df2["patient_id"].isin(train_patients)].copy()
test_df2 = model_df2[model_df2["patient_id"].isin(test_patients)].copy()

print("Train rows:", len(train_df2))
print("Test rows:", len(test_df2))
print("Train patients:", train_df2["patient_id"].nunique())
print("Test patients:", test_df2["patient_id"].nunique())
```

```python
X_train2 = train_df2.drop(columns=["HR", "patient_id"])
y_train2 = train_df2["HR"]

X_test2 = test_df2.drop(columns=["HR", "patient_id"])
y_test2 = test_df2["HR"]
```

```python
categorical_cols2 = ["sex", "study", "condition_group"]
numeric_cols2 = [c for c in X_train2.columns if c not in categorical_cols2]

X_train2 = pd.get_dummies(X_train2, columns=categorical_cols2)
X_test2 = pd.get_dummies(X_test2, columns=categorical_cols2)

X_train2, X_test2 = X_train2.align(X_test2, join="left", axis=1, fill_value=0)

scaler2 = StandardScaler()
X_train2[numeric_cols2] = scaler2.fit_transform(X_train2[numeric_cols2])
X_test2[numeric_cols2] = scaler2.transform(X_test2[numeric_cols2])
```

```python
from sklearn.preprocessing import StandardScaler

y_scaler = StandardScaler()

y_train2_scaled = y_scaler.fit_transform(y_train2.values.reshape(-1, 1))
y_test2_scaled = y_scaler.transform(y_test2.values.reshape(-1, 1))
```

```python
X_train2 = X_train2.astype(np.float32)
X_test2 = X_test2.astype(np.float32)

X_train2_t = torch.tensor(X_train2.to_numpy(), dtype=torch.float32)
X_test2_t = torch.tensor(X_test2.to_numpy(), dtype=torch.float32)

y_train2_t = torch.tensor(y_train2_scaled, dtype=torch.float32)
y_test2_t = torch.tensor(y_test2_scaled, dtype=torch.float32)
```

```python
class HRPredictorV2(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(32, 16),
            nn.ReLU(),

            nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.model(x)

model2 = HRPredictorV2(X_train2_t.shape[1])
```

```python
criterion2 = nn.SmoothL1Loss()
optimizer2 = optim.Adam(model2.parameters(), lr=0.001)
```

```python
epochs = 300

for epoch in range(epochs):
    model2.train()

    preds2 = model2(X_train2_t)
    loss2 = criterion2(preds2, y_train2_t)

    optimizer2.zero_grad()
    loss2.backward()
    optimizer2.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch} | Loss {loss2.item():.4f}")
```

```python
model2.eval()

with torch.no_grad():
    test_preds2 = model2(X_test2_t).cpu().numpy()

# inverse transform back to bpm
test_preds2_hr = y_scaler.inverse_transform(test_preds2).flatten()
y_test2_hr = y_test2.values

rmse2 = np.sqrt(np.mean((test_preds2_hr - y_test2_hr) ** 2))
mae2 = np.mean(np.abs(test_preds2_hr - y_test2_hr))

print("Model 2 RMSE:", rmse2)
print("Model 2 MAE:", mae2)
```

```python
features3 = [
    "sex",
    "age",
    "BMI",
    "BSA",
    "study",
    "condition_group",
    "watts",
    "MAP",
    "SBP",
    "DBP",
    "CO",
    "SV",
    "VO2Lmin",
    "VCO2",
    "RR"
]

target = "HR"

model_df3 = df[features3 + [target, "patient_id"]].copy()

model_df3.head()
```

```python
for col in features3 + [target]:
    model_df3[col] = pd.to_numeric(model_df3[col], errors="coerce")

model_df3["sex"] = model_df3["sex"].replace(".", np.nan)
```

```python
model_df3.isna().sum()
```

```python
# Fix study
model_df3["study"] = df.loc[model_df3.index, "study"]

# Fix condition_group
model_df3["condition_group"] = df.loc[model_df3.index, "condition_group"]

# Check again
model_df3[["study", "condition_group"]].isna().sum()
```

```python
model_df3 = model_df3.drop(columns=["CO", "SV"])
```

```python
model_df3 = model_df3.dropna(subset=["HR"])
```

```python
# sex
model_df3["sex"] = model_df3["sex"].fillna(model_df3["sex"].mode()[0])

# watts
model_df3["watts"] = model_df3["watts"].fillna(model_df3["watts"].median())

# pressure columns
for col in ["MAP", "SBP", "DBP"]:
    model_df3[col] = model_df3[col].fillna(model_df3[col].median())

# oxygen
for col in ["VO2Lmin", "VCO2"]:
    model_df3[col] = model_df3[col].fillna(model_df3[col].median())

# RR (higher missing → median is safest for now)
model_df3["RR"] = model_df3["RR"].fillna(model_df3["RR"].median())
```

```python
model_df3.isna().sum()
```

```python
train_df3 = model_df3[model_df3["patient_id"].isin(train_patients)].copy()
test_df3 = model_df3[model_df3["patient_id"].isin(test_patients)].copy()

print("Train rows:", len(train_df3))
print("Test rows:", len(test_df3))
print("Train patients:", train_df3["patient_id"].nunique())
print("Test patients:", test_df3["patient_id"].nunique())
```

```python
X_train3 = train_df3.drop(columns=["HR", "patient_id"])
y_train3 = train_df3["HR"]

X_test3 = test_df3.drop(columns=["HR", "patient_id"])
y_test3 = test_df3["HR"]
```

```python
categorical_cols3 = ["sex", "study", "condition_group"]
numeric_cols3 = [c for c in X_train3.columns if c not in categorical_cols3]

X_train3 = pd.get_dummies(X_train3, columns=categorical_cols3)
X_test3 = pd.get_dummies(X_test3, columns=categorical_cols3)

X_train3, X_test3 = X_train3.align(X_test3, join="left", axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler3 = StandardScaler()

X_train3[numeric_cols3] = scaler3.fit_transform(X_train3[numeric_cols3])
X_test3[numeric_cols3] = scaler3.transform(X_test3[numeric_cols3])
```

```python
y_scaler3 = StandardScaler()

y_train3_scaled = y_scaler3.fit_transform(y_train3.values.reshape(-1,1))
y_test3_scaled = y_scaler3.transform(y_test3.values.reshape(-1,1))
```

```python
X_train3 = X_train3.astype(np.float32)
X_test3 = X_test3.astype(np.float32)

X_train3_t = torch.tensor(X_train3.to_numpy(), dtype=torch.float32)
X_test3_t = torch.tensor(X_test3.to_numpy(), dtype=torch.float32)

y_train3_t = torch.tensor(y_train3_scaled, dtype=torch.float32)
y_test3_t = torch.tensor(y_test3_scaled, dtype=torch.float32)
```

```python
class HRPredictorV3(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, 16),
            nn.ReLU(),

            nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.model(x)

model3 = HRPredictorV3(X_train3_t.shape[1])
```

```python
criterion3 = nn.SmoothL1Loss()
optimizer3 = optim.Adam(model3.parameters(), lr=0.001)
```

```python
epochs = 300

for epoch in range(epochs):
    model3.train()

    preds3 = model3(X_train3_t)
    loss3 = criterion3(preds3, y_train3_t)

    optimizer3.zero_grad()
    loss3.backward()
    optimizer3.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch} | Loss {loss3.item():.4f}")
```

```python
model3.eval()

with torch.no_grad():
    test_preds3 = model3(X_test3_t).cpu().numpy()

# convert back to bpm
test_preds3_hr = y_scaler3.inverse_transform(test_preds3).flatten()
y_test3_hr = y_test3.values

rmse3 = np.sqrt(np.mean((test_preds3_hr - y_test3_hr) ** 2))
mae3 = np.mean(np.abs(test_preds3_hr - y_test3_hr))

print("Model 3 RMSE:", rmse3)
print("Model 3 MAE:", mae3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert to numpy arrays just in case
actual_hr_ml = np.array(y_test3_hr)
pred_hr_ml = np.array(test_preds3_hr)

# Metrics
rmse_ml = np.sqrt(np.mean((pred_hr_ml - actual_hr_ml) ** 2))
mae_ml = np.mean(np.abs(pred_hr_ml - actual_hr_ml))

# Plot
plt.figure(figsize=(7, 7))
plt.scatter(actual_hr_ml, pred_hr_ml, alpha=0.75)

# 45-degree reference line
min_val = min(actual_hr_ml.min(), pred_hr_ml.min())
max_val = max(actual_hr_ml.max(), pred_hr_ml.max())
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--")

plt.xlabel("Actual HR (bpm)")
plt.ylabel("Predicted HR (bpm)")
plt.title(f"Model 3: Predicted vs Actual HR\nRMSE = {rmse_ml:.2f}, MAE = {mae_ml:.2f}")
plt.grid(True)
plt.show()
```
